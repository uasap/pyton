```python
# Построение ящика с усами для квадратуры жилой площади
df_house_data = pd.read_csv("kc-house-data.csv")
plt.boxplot(df_house_data['sqft_living'])
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Квадратные футы')
plt.title(' для квадратуры жилой площади')
plt.show()
```

```python
import numpy as np
# Построение ящика с усами для логарифмированной квадратуры жилой площади
plt.boxplot(np.log(df_house_data['sqft_living']))
plt.xlabel('Квадратура жилой площади (логарифм)')
plt.ylabel('Логарифм квадратных футов')
plt.title('Ящик с усами для логарифма квадратуры  площади')
plt.show()
```

```python
# Найти самую минимальную цену на недвижимость
min_price = df_house_data['price'].min()
# Найти самую минимальную площадь
min_sqft_living = df_house_data['sqft_living'].min()
print("Самая минмальная цена на недвижимость:", min_price)
print("Самая минимальная площадь:", min_sqft_living)
```

```python
# Фильтрация данных
filtered_data = df_house_data[(df_house_data['yr_built'] > 2000) & (df_house_data['sqft_living'] > 0)]
cheapest_property = filtered_data[filtered_data['price'] == filtered_data['price'].min()]
print("Самая дешевая недвижимость, построенная после 2000 года  0:")
print(cheapest_property[['price', 'yr_built', 'sqft_living']])
```

```python
# Построение гистограммы года постройки
plt.hist(df_house_data['yr_built'], bins=30)
plt.xlabel('Год постройки')
plt.ylabel('Количество наблюдений')
plt.title('Год  постройки')
plt.show()
```

```python
# Построение "Ящика с усами" года постройки
plt.boxplot(df_house_data['yr_built'])
plt.xlabel('Год постройки')
plt.ylabel('Количество наблюдений')
plt.title('Распределение года постройки')
plt.show()
```

```python
#Условие 2: 2 задача

import pandas as pd
import matplotlib.pyplot as plt

# Построение круговой диаграммы для распределения домов от наличия вида на набережную
df_house_data = pd.read_csv("kc-house-data.csv")
waterfront_counts = df_house_data['waterfront'].value_counts()
plt.figure(figsize=(2, 2))
plt.pie(waterfront_counts, labels=['Нет', 'Да'], autopct='%1.1f%%', startangle=90)
plt.title('Расприделение домов от наличия вида на набережную')
plt.axis('equal')  #  масштаб по осям x и y для круговой диаграммы
plt.show()
```

```python
# Построение гистограммы для распределения этажей домов
df_house_data['floors'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Этажи ')
plt.show()
```

```python
import seaborn as sns
# Установка стиля Seaborn 
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.countplot(data=df_house_data, x='condition', hue='floors')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Многорядная столбчатая диаграмма для состояния и этажей')
plt.legend(title='Количество этажей')
plt.show()
```

```python
# Зависимость стоимости от квадратуры жилой площади (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_house_data, x='sqft_living', y='price', alpha=0.7)
plt.xlabel('Квадратура жилой площади (sqft)')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Зависимость стоимости от квадратуры жилой площади')
plt.show()
```

```python
#  Распределение стоимости недвижимости по количеству комнат 
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_house_data, x='bedrooms', y='price')
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Стоимость недвижимости по количеству комнат')
plt.show()
```
